```python
import numpy as np
import matplotlib.pyplot as plt
```

# Численное решение нелинейного уравнения. Метод Ньютона

Исходная функция

```python
def f(x):
    return pow(np.abs(np.tan(x)), x) * np.sin(x) + 1 / 2
```

График исходной функции

```python
a = -2
b = 1
plt.plot(np.linspace(a, b), f(np.linspace(a, b)))
plt.grid(True)
```

Метод Ньютона с разростной аппроксимацией производной

Желаемая ошибка

```python
n = 8
err_adm = 2 ** (-n)
err = [np.inf]
```

Границы поиска

```python
a = -2
b = 1
x = [0]
```

Итерационный процесс

```python
h = 0.01 * (b - a)
k = 1

while abs(err[k - 1]) > err_adm:
    f_derr = (f(x[k - 1] + h) - f(x[k - 1] - h)) / (2 * h)
    x.append(x[k - 1] - f(x[k - 1]) / f_derr)
    err.append(- f(x[k - 1]) / f_derr)
    k += 1
```

Затраченное количество итераций

```python
k
```

Приближенное решение

```python
f(x[k - 1])
```

График зависимости ошибки от количества итераций

```python
plt.plot(np.arange(k), np.abs(err))
plt.grid(True)
```

Метод Ньютона с постоянным значением производной

Желаемая ошибка

```python
n = 8
err_adm = 2 ** (-n)
err = [np.inf]
```

Границы поиска

```python
a = -2
b = 1
x = [0]
```

Разностная производная, вычисленная в точке х0

```python
h = 0.01 * (b - a)
f_derr = (f(x[0] + h) - f(x[0] - h)) / (2 * h)
```

Итерационный процесс

```python
k = 1

while abs(err[k - 1]) > err_adm:
    x.append(x[k - 1] - f(x[k - 1]) / f_derr)
    err.append(- f(x[k - 1]) / f_derr)
    k += 1
```

Затраченное количество итераций

```python
k
```

Приближенное решение

```python
f(x[k - 1])
```

График зависимости ошибки от количества итераций

```python
plt.plot(np.arange(k), np.abs(err))
plt.grid(True)
```
